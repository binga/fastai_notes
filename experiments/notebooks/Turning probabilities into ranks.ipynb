```python
import numpy as np
import pandas as pd
preds = pd.DataFrame({'id': range(3), 
                      'class1': [0.1, 0.4, 0.5], 
                      'class2': [0.5, 0.35, 0.1], 
                      'class3': [0.2, 0.25, 0.55]})
preds = preds[['id', 'class1', 'class2', 'class3']]
preds.head()
```

```python
preds_melt = preds.melt(id_vars='id', value_vars=['class1', 'class2', 'class3'])
preds_melt
```

```python
preds_melt['rank'] = preds_melt.groupby('id')['value'].rank(ascending=False)
preds_melt
```

```python
preds_ranks = preds_melt.pivot_table(index='id', columns='variable', values='rank').reset_index()
preds_ranks
```
